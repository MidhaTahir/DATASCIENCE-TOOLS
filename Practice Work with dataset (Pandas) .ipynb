```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
index=['Utah', 'Ohio', 'Texas', 'Oregon'])
```

```python
frame
```

```python
# np.abs(frame)
```

```python
frame.max()
```

```python
#lamba fn 
f1 = lambda x: x.max() - x.min()
frame.apply(f1)
```

```python
f2 = lambda x: sum(x)
frame.apply(f2)
```

```python
f3 = lambda y: y**2
frame.apply(f3)
```

```python
f = lambda x: x.max() - x.min()
frame.apply(f, axis='columns') #row axis is 0 , column axis 1 by default --> 0 
```

```python
f = lambda x: x.max() - x.min()
frame.apply(f, axis=1)
```

```python
frame.max(axis=1)
```

# SORTING AND RANKING

```python
obj = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
obj
```

```python
obj.sort_index()
```

```python
obj.sort_index(ascending=False)
obj
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),index=['three', 'one'],columns=['d', 'a', 'b', 'c'])
frame
```

```python
frame.sort_index()
```

```python
frame.sort_index(axis=1)
```

```python
fr = pd.DataFrame(np.arange(12,0,-1).reshape((3, 4)),index=['three', 'one','four'],columns=['d', 'a', 'b', 'c'])
fr
```

```python
fr.sort_values(by='a',inplace=True) #a is sorted
fr
```

```python
fr['d']['one'] = 2
fr
```

```python
fr.sort_values(by='d')
```
